```python
#############################################################EXPERIENCE##############################################################
def experiencefunc(exp_list,la):
    #la in place of 5 for all experience
    for i in range(0,5):
        try:
            #exp_list = driver.find_element_by_css_selector('ul.expandable-list__list')
            a = exp_list.find_elements_by_tag_name("li")[i]
            abcd= a.find_element_by_css_selector("dl.background-entity__summary")
            #####################################################################COMPANY
            try:
                try:
                    company= abcd.find_element_by_css_selector("a[class='position-item__company-link']")
                    compv=company.get_attribute("innerHTML")
                except NoSuchElementException:
                    company= abcd.find_element_by_css_selector("dd[class='background-entity__summary-definition--subtitle']")
                    compv=company.get_attribute("innerHTML")
            except NoSuchElementException:
                compv="NA"    
            ####################################################################DURATION
            try:
                duration= abcd.find_element_by_css_selector("span[class='background-entity__duration']")
                durv=duration.get_attribute("innerHTML")
            except NoSuchElementException:
                durv="NA"
            ####################################################################SUMMARY
            try:
                summary= abcd.find_element_by_css_selector("dd[class='background-entity__summary-definition--description']")
                sumv=summary.get_attribute("innerHTML")
            except NoSuchElementException:
                sumv="NA"
            ####################################################################SUMMARY
            try:
                location= abcd.find_element_by_css_selector("dd[class='background-entity__summary-definition--location']")
                locv=location.get_attribute("innerHTML")
            except NoSuchElementException:
                locv="NA"
            ####################################################################POSITION
            try:
                position= a.find_element_by_css_selector("em.sh")
                posv=position.get_attribute("innerHTML")
            except NoSuchElementException:
                position= a.find_element_by_css_selector("a[class='position-item__position-title-link ember-view']")
                posv=position.get_attribute("innerHTML")
            pumv=durv.strip() +", "+ sumv.strip()
            #print(pumv)
            experiencelist={}
            experiencelist["Company:"]=compv.strip()
            experiencelist["Position:"]=posv.strip()
            #experiencelist["Duration:"]=durv.strip()
            experiencelist["Location:"]=locv.strip()
            experiencelist["Duration & Summary:"]=pumv.strip()
#             print(experiencelist)
            
            experience[i+1]=experiencelist
#             print("Company:",compv.strip())  
#             print("Duration:",durv.strip())
#             print("Position:",posv.strip())
#             print("Summary:",sumv.strip())
#             print(experience)
        except:
            break
    return experience
```

```python
###########################################################SKILSS###################################
def skillsfunc(skills_list,sle):
    for i in range(0,sle):
        try:
            #exp_list = driver.find_element_by_css_selector('ul.expandable-list__list')
            a = skills_list.find_elements_by_tag_name("li")[i]
            #print(a)
            abcd= a.find_element_by_css_selector("dl.skill__wrapper")
            #####################################################################SKILL
            try:
                company= abcd.find_element_by_css_selector("dt[class='skill__skill-name']")
                compv=company.get_attribute("innerHTML")
            except NoSuchElementException:
                compv="NA"    
            skilllist.append(compv.strip())
            #print(skilllist)
        except NoSuchElementException:
            break
    return skilllist
```
